Model Validation

Now we have a model which predict certain output. But have Accurate and correct predict does model is doing. For that we perform model validation.

The relevent measure of model quality is predictive accuracy. In other words does model predition is close to what actually happen.

many people make mistake in predictive accuracy. they make prediction with training data and compare with targeted value in training data.

TO solve this problem we summarize the model quality into understandable way.

we cant compare predicted values and actual values each of 10000 houses. So we do that in a single metrics.

There are many metrics for summarize the model quality. Here we use Mean Absolute Error(MAE)

here error = actual - predicted

Absolute Error refers to difference between prediction value and actual target value in training data.

MAE takes the average of absolute error. And this MAE value is the model quality score.

```python
# lets find the model quality score with the training data
from sklearn.metrics import mean_absolute_error
predic = mel_model.predict(X)

mae = mean_absolute_error(y, predic) #measuring the quality of model
mae
```

```python
#the above measure is "in-sample". We used same data for model training and model validation. Its a bad practise.
#it give accurate prediction when measure with the same trained data. But when its face with new data the model generate
#inaccurate predictions
# for that we split the data into two parts 1.Training data 2.Validation data.
# we split the the input data i.e X as train_X & val_X and target value i.e y as train_y & val_y
#using train_test_split from model_selection
```

```python
from sklearn.model_selection import train_test_split

#spliting the data
train_X, val_X, train_y, val_y = train_test_split(X,y,random_state=0)

#fitting into the model
mel_model.fit(train_X,train_y)

#making the prediction
val_prediction = mel_model.predict(val_X)
count = 0
for i in val_prediction:
  count+=1

print("count = ",count)

#measuring
print(mean_absolute_error(val_y, val_prediction))
```

```python
# 1st with row data
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
data = pd.read_csv('/content/melb_data.csv')

data = data.dropna()

y = data.Price

features = ['Rooms','Bathroom','Landsize','Lattitude','Longtitude']

X = data[features]

# building model with DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=1)

#fitting model with actual data
dt_model.fit(X,y)
dt_predic1 = dt_model.predict(X)
mae1 = mean_absolute_error(dt_predic1,y)
print("Decision Tree Regressor")
print("Mean Absolute Error for actual data:")
print(mae1)

#now split the data into train_X, val_x, train_y, val_y
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state=1)

# fitting the model with split data for DesicionTreeRegressor
dt_model.fit(train_X,train_y)
dt_predic2 = dt_model.predict(val_X)
mae2 = mean_absolute_error(dt_predic2,val_y)
print("Mean Absolute Error for split data:")
print(mae2)


# lets built model with another method with RandomForestRegressor
rf_model = RandomForestRegressor(random_state = 1)

rf_model.fit(X,y)
rf_predic1 = rf_model.predict(X)

# mean absolute error
mae3 = mean_absolute_error(rf_predic1, y)
print("\nRandomForestRegressor")
print("Mean Absoulte Error for Raw data : ")
print(mae3)

#lets fit with split data
rf_model.fit(train_X,train_y)
rf_predic2 = rf_model.predict(val_X)

# absolute error
mae4 = mean_absolute_error(rf_predic2, val_y)
print("Mean Absoulte Error for split data : ")
print(mae4)
```

From now handling missing values

There are three approches to handle missing values

1. dropping columns which have missing values.
2. Imputaion - fill with some number.
3. Extension for Imputaion - create a new column which locate the impute entries.

```python
data2 = pd.read_csv('/content/melb_data.csv')
```

```python
y2 = data2.Price

# To keep things simple, we'll use only numerical predictors
melb_predictors = data2.drop(['Price'], axis=1)
X2 = melb_predictors.select_dtypes(exclude=['object'])

# Divide data into training and validation subsets
X2_train, X2_val, y2_train, y2_val = train_test_split(X2, y2, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
# def which return the mean absolute error of model
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

def model_score(X2_train,X2_val,y2_train,y2_val):
  model = RandomForestRegressor(n_estimators=10, random_state=0)
  model.fit(X2_train,y2_train)
  preds = model.predict(X2_val)
  return mean_absolute_error(y2_val, preds)
```

```python
# Approach 1
col_with_missing = [col for col in X2_train.columns
                    if X2_train[col].isnull().any()]

print(col_with_missing)

#dropping columns in training and validation data
drop_X2_train = X2_train.drop(col_with_missing,axis=1)
drop_X2_val = X2_val.drop(col_with_missing,axis=1)

print("MAE with removing columns with missing data : ")
print(model_score(drop_X2_train,drop_X2_val,y2_train,y2_val))
```

```python
# Approach 2
# imputation here we fill the nan vallues with mean of existing values
from sklearn.impute import SimpleImputer

my_impute = SimpleImputer()

imp_X2_train = pd.DataFrame(my_impute.fit_transform(X2_train))
imp_X2_val = pd.DataFrame(my_impute.transform(X2_val))

# By performing imputation it remove column names then add again
imp_X2_train.columns = X2_train.columns
imp_X2_val.columns = X2_val.columns

print("Approach 2 (Imputation): ")
print(model_score(imp_X2_train,imp_X2_val,y2_train,y2_val))
```

```python
# Approach 3
# we perfrom imputation and we need to track were impute as happen

# make a copy to avoid changing original data
X2_train_2 = X2_train.copy()
X2_val_2 = X2_val.copy()

#making new columns indicating what will be imputed
for col in col_with_missing:
  imp_X2_train_2 = pd.DataFrame(my_impute.fit_transform(X2_train_2))
  imp_X2_val_2 = pd.DataFrame(my_impute.transform(X2_val_2))

# imputation remove column name add back
imp_X2_train_2.columns = X2_train_2.columns
imp_X2_val_2.columns = X2_val_2.columns

print("MAE for Approach 3 (en extention to imputation):")
print(model_score(imp_X2_train_2,imp_X2_val_2,y2_train,y2_val))
```
